**Problem**
to predict whether student if studies more hour will get more marks?


**Solution** 
plot of a study hours and training a model based on study hours per week and understan its relationship between final score

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

#dataset
data = pd.read_csv('student_performance_dataset.csv')

#input and output features
x = data[['Study_Hours_per_Week']]
y= data['Final_Exam_Score']

#train the model
model = LinearRegression()
model.fit(x, y)
predicted_scores = model.predict(x)

#valid regression matrix
mae = mean_absolute_error(y, predicted_scores)
mse = mean_squared_error(y, predicted_scores)
r2 = r2_score(y, predicted_scores)
rmse = np.sqrt(mse)

#show results
print(f'Mean Absolute Error: {round(mae,2)}')
print(f'Mean Squared Error: {round(mse,2)}')
print(f'R^2 Score: {round(r2,2)}')
print(f'Root Mean Squared Error: {round(rmse,2)}')

#histogram
plt.figure(figsize=(10,2))
plt.hist(data["Final_Exam_Score"], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Final Exam Scores')
plt.xlabel('Final Exam Score')
plt.ylabel('Number Of Students')
plt.grid(True)
plt.show() 

#scatter + regression line
plt.figure(figsize=(10,2))
plt.scatter(x,y, color='skyblue', edgecolor='black', label='Actual Scores')
plt.plot(x, predicted_scores, color='red', linewidth=2, label='predicted scores (Regression Line)')
plt.title('Model Prediction VS Actual Scores')
plt.xlabel('Study Hours Per Week')
plt.ylabel('Final Output Score')
plt.grid(True)
plt.show() 

#
new_hours = 9
predict_new_score = model.predict([[new_hours]])
print(f'Predicted Final Exam Score for {new_hours} Study Hours per Week: {round(predict_new_score[0],2)}')
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

#dataset
data = pd.read_csv('student_performance_dataset.csv')

#input and output features
x = data[['Study_Hours_per_Week']]
y= data['Final_Exam_Score']
z= data['Past_Exam_Scores']

#train the model
model = LinearRegression()
model.fit(x, y, z)
predicted_scores = model.predict(x)

#valid regression matrix
mae = mean_absolute_error(y, predicted_scores)
mse = mean_squared_error(y, predicted_scores)
r2 = r2_score(y, predicted_scores)
rmse = np.sqrt(mse)

#show results
print(f'Mean Absolute Error: {round(mae,2)}')
print(f'Mean Squared Error: {round(mse,2)}')
print(f'R^2 Score: {round(r2,2)}')
print(f'Root Mean Squared Error: {round(rmse,2)}')

#histogram
plt.figure(figsize=(10,2))
plt.hist(data["Final_Exam_Score"], bins=30, color='skyblue', edgecolor='black')
plt.hist(data['Past_Exam_Scores'], bins=30, color='blue', edgecolor='black')
plt.title('Distribution of Final Exam Scores')
plt.xlabel('Final Exam Score')
plt.ylabel('Number Of Students')
plt.grid(True)
plt.show() 

#scatter + regression line
plt.figure(figsize=(10,2))
plt.scatter(x,y, color='skyblue', edgecolor='black', label='Actual Scores')
plt.scatter(x,z, color='blue', edgecolor='black', label='Actual Scores')
plt.plot(x, predicted_scores, color='red', linewidth=2, label='predicted scores (Regression Line)')
plt.title('Model Prediction VS Actual Scores')
plt.xlabel('Study Hours Per Week')
plt.ylabel('Final Output Score')
plt.grid(True)
plt.show() 

#
new_hours = 9
predict_new_score = model.predict([[new_hours]])
print(f'Predicted Final Exam Score for {new_hours} Study Hours per Week: {round(predict_new_score[0],2)}')
```
